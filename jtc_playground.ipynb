```python
from facenet_pytorch import MTCNN, InceptionResnetV1

# Load and initialize objects from GitHub repo

# If required, create a face detection pipeline using MTCNN:
mtcnn = MTCNN(image_size=160, margin=0)

# Create an inception resnet (in eval mode):
resnet = InceptionResnetV1(pretrained='vggface2').eval()

```

```python
from PIL import Image
from os.path import join
import matplotlib.pyplot as plt

# Load training image
train_img_path = r"C:\Users\jjuus\OneDrive - purdue.edu\ECE 50024\Kaggle\train_small\train_small"
train_img1 = Image.open(join(train_img_path, '1.jpg'))
display(train_img1)

# Get cropped and prewhitened image tensor
train_img1_cropped = mtcnn(train_img1, save_path="./test.jpg")

# Or, if using for VGGFace2 classification
resnet.classify = True
img_probs = resnet(train_img1_cropped.unsqueeze(0))
print("img_probs: ", img_probs)
print("img_probs.shape: ", img_probs.shape)

```
